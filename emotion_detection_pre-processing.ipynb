```python
import pandas as pd
import os
```

## Pre-Processing

```python
# bath run of all the processed images indivdual csv file into one with a class label
def create_emotions_csv(filepath, emotion):
    path = filepath
    extension = ".csv"
    files = [file for file in os.listdir(path) if file.endswith(extension)]
    dfs = []
    for file in files:
        df = pd.read_csv(os.path.join(path, file))
        dfs.append(df)
    csvname = pd.concat(dfs, ignore_index=True)
    csvname["Emotion Detected"] = emotion
    return csvname
    
```

```python
anger_processed = create_emotions_csv("/Users/sanzi/Desktop/Facial-Recognition-Music-Recommendation-System/fer_ckplus_kdef/anger/processed/", 
                    "Anger")
contempt_processed = create_emotions_csv("/Users/sanzi/Desktop/Facial-Recognition-Music-Recommendation-System/fer_ckplus_kdef/contempt/processed/", 
                    "Contempt")
disgust_processed = create_emotions_csv("/Users/sanzi/Desktop/Facial-Recognition-Music-Recommendation-System/fer_ckplus_kdef/disgust/processed/", 
                    "Disgust")
fear_processed = create_emotions_csv("/Users/sanzi/Desktop/Facial-Recognition-Music-Recommendation-System/fer_ckplus_kdef/fear/processed/", 
                    "Fear")
```
