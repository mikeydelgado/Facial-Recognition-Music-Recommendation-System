## Training Models

```python
# Metircs on how well the models performed
def metics_reports(y_label_prediction, y_label_training):
    
    # Accuracy Score Prediciton 
    print("Accurary Score")
    print(accuracy_score(y_label_training, y_label_prediction))
    
    # Confusion Matrix Prediciton 
    print("Confusin Matrix Score")
    print(confusion_matrix(y_label_training, y_label_prediction))

    # precision_recall_fscore_support Prediciton 
    print("Precision Recall Score")
    print(precision_recall_fscore_support(y_label_training, y_label_prediction))
    
```

```python
X_train_intensity, X_test_intensity, y_train_intensity, y_test_intensity = train_test_split(emotion_training_intensity, emotion_training_labels, test_size=0.33, random_state=42)
```

```python
svm_model = svm.SVC()
svm_model.fit(X_train_intensity, y_train_intensity)
svm_prediction = svm_model.predict(X_test_intensity)
```

```python
metics_reports(y_test_intensity,svm_prediction)
```
